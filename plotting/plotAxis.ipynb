```python
from mayavi import mlab
mlab.init_notebook('x3d',800,800)
import sys
sys.path.append("..")
```

```python
import numpy as np
import mayavi as maya
import focusadd
from numpy import cos, sin
```

```python
from focusadd.surface.readAxis import read_axis
from focusadd.surface.Axis import Axis
```

```python
def plot_axis_and_TNB(num_zeta, res, sk, color=(0.8, 0.0, 0.0)):
    axis = Axis(read_axis("../focusadd/initFiles/axes/ellipticalAxis4Rotate.txt"), num_zeta, res = res)
    x,y,z = axis.get_xyz()
    T = axis.get_tangent()
    N = axis.get_normal()
    B = axis.get_binormal()
    #a = mlab.plot3d(x, y, z, np.sin(axis.get_zeta()), tube_radius=0.02, line_width = 0.01, color=color)
    a = mlab.quiver3d(x[::sk],y[::sk],z[::sk],T[::sk,0],T[::sk,1],T[::sk,2], color=color)
    a = mlab.quiver3d(x[::sk],y[::sk],z[::sk],N[::sk,0],N[::sk,1],N[::sk,2], color=color)
    a = mlab.quiver3d(x[::sk],y[::sk],z[::sk],B[::sk,0],B[::sk,1],B[::sk,2], color=color)
    return a
```

```python
mlab.clf()
a = plot_axis_and_TNB(8, 1, 1)
b = plot_axis_and_TNB(128, 1, 16, color=(0.0,0.0,0.8))
a
b
```
