```python
from mayavi import mlab
mlab.init_notebook()
import sys
sys.path.append("..")
```

```python
import numpy as np
import mayavi as maya
import focusadd
from numpy import cos, sin
```

```python
import focusadd.surface.readAxis as readAxis
```

```python
def plot_axis():
    axis, _, _, _, _ = readAxis.readAxis("../focusadd/initFiles/axes/defaultAxis.txt", 64)
    x,y,z = axis.get_xyz()
    l = mlab.plot3d(x, y, z, np.sin(axis.get_zeta()), tube_radius=0.02, line_width = 0.01, colormap="Spectral")
    return l
    
```

```python
mlab.clf()
l = plot_axis()
l
```

```python
mlab.savefig("defaultAxis.x3d", size=(1000,1000), figure=mlab.gcf())
```
